## Gradient Descent with Exact Line Search

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

$f:(x_1,x_2) \mapsto (x_1-2)^2+2(x_2+1)^2+1$

$ \nabla f(x_1,x_2) = \begin{bmatrix} 2(x_1-2)\\ 4(x_2+1)\end{bmatrix}= \begin{bmatrix} 2x_1-4\\4x_2+4 \end{bmatrix}$

$H_f(x_1,x_2)=\begin{bmatrix} 2 & 0\\0 & 4\end{bmatrix}$

```python
X_1 = np.arange(-4,4,0.1)
X_2 = np.arange(-4,4,0.1)
X_1,X_2 = np.meshgrid(x_1,x_2)
```

```python
def f(x_1,x_2):
    return (x_1-2)**2 + 2*(x_2+1)**2 +1
def grad_f(x_1,x_2):
    return np.array([(2*x_1-4),(4*x_2+4)])
```

```python
X_3 = f(X_1,X_2)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_1,X_2,X_3,cmap=cm.viridis)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("f(x1, x2)")
plt.show()
```

```python
plt.close()
fig,ax = plt.subplots()
ax.set_aspect(aspect=1)
CS=ax.contour(X_1,X_2,X_3,levels=[2,4,6,8,10,15,20,25,30,35,40])
ax.clabel(CS,inline=1,fontsize=10)
ax.plot(2,-1,"o")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.plot([-4,2,2],[-1,-1,-4],"--");
```

## Start Gradient Descent
### Step 1
Starting point $x_0=(-2,2)$

```python
f(-2,2)
```

```python
plt.close()
fig,ax = plt.subplots()
ax.set_aspect(aspect=1)
CS=ax.contour(X_1,X_2,X_3,levels=[2,4,6,8,10,15,20,25,30,35,40])
ax.clabel(CS,inline=1,fontsize=10)
ax.plot(2,-1,"o")
ax.plot(2,2,"o")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.plot([-4,2,2],[-1,-1,-4],"--");
```

```python
-grad_f(-2,2)
```

```python
plt.close()
fig,ax = plt.subplots()
ax.set_aspect(aspect=1)
CS=ax.contour(X_1,X_2,X_3,levels=[2,4,6,8,10,15,20,25,30,35,40])
ax.clabel(CS,inline=1,fontsize=10)
ax.plot(2,-1,"o")
ax.plot([-2,2],[2,-4],"o--",markevery=[0])
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.plot([-4,2,2],[-1,-1,-4],"--");
```

### Here we compute the step size $\lambda$ via line search


$x_1=\begin{bmatrix} -2\\ 2\end{bmatrix}+\lambda\begin{bmatrix} 8\\ -12\end{bmatrix}$

$g(\lambda)=f(x_0+\lambda v)=f\Big(\begin{bmatrix} -2\\ 2\end{bmatrix}+ \lambda \begin{bmatrix} 8\\ -12\end{bmatrix}\Big)=f(-2+8\lambda,2-12\lambda)$

$x_1=\begin{bmatrix} -2\\ 2\end{bmatrix} + 0.3 \begin{bmatrix} 8\\ -12\end{bmatrix} \approx \begin{bmatrix} 0.4\\ -1.6\end{bmatrix}$

```python
plt.close()
fig,ax = plt.subplots()
ax.set_aspect(aspect=1)
CS=ax.contour(X_1,X_2,X_3,levels=[2,4,6,8,10,15,20,25,30,35,40])
ax.clabel(CS,inline=1,fontsize=10)
ax.plot(2,-1,"o")
ax.plot([-2,0.4,2],[2,-1.6,-4],"o--",markevery=[0,1]);
ax.set_xlabel("x1")
ax.set_ylabel("x2")
```
