$ \nabla f(x_1,x_2) = \begin{bmatrix} 2(x_1-2)\\ 4(x_2+1)\end{bmatrix}= \begin{bmatrix} 2x_1-4\\4x_2+4 \end{bmatrix}$

$H_f(x_1,x_2)=\begin{bmatrix} 2 & 0\\0 & 4\end{bmatrix}$

```python
X_1 = np.arange(-4,4,0.1)
X_2 = np.arange(-4,4,0.1)
X_1,X_2 = np.meshgrid(x_1,x_2)
```

## Example 2: Higher Order Function:

$f:(x_1,x_2) \mapsto (x_1-2)^2+2(x_1+1)^2(x_2+1)^2+(x_2-1^4)+1$

$ \nabla f(x_1,x_2) = \begin{bmatrix} 4(x_1-2)^3+4(x_1+1)(x_2+1)^2 \\ 4(x_1+1)^2(x_2+1)+4(x_2-1)^3\end{bmatrix}$

$H_f(x_1,x_2)=\begin{bmatrix} 12(x_1-2)^2+4(x_2+1)^2 & 8(x_2+1)(x_1+1)\\8(x_2+1)(x_1+1) & 12(x_2-1)^2\end{bmatrix}$

```python
def f(x_1, x_2):
    return (x_1 - 2) ** 2 + 2 * (x_1 + 1) ** 2 * (x_2 + 1) ** 2 + (x_2 - 1) ** 4 + 1

def grad_f(x_1, x_2):
    return np.array([(4 * (x_1 - 2) ** 3 + 4 * (x_1 + 1) * (x_2 + 1) ** 2),
                     (4 * (x_1 + 1) ** 2 * (x_2 + 1) + 4 * (x_2 - 1) ** 3)])

def hessian_f(x_1, x_2):
    return np.array([(12 * (x_1 - 2) ** 2 + 4 * (x_2 + 1) ** 2, 8 * (x_1 + 1) * (x_2 + 1)),
                     (8 * (x_1 + 1) * (x_2 + 1), 12 * (x_2 - 1) ** 2)])
```

```python
x_1=np.arange(-3,3,0.1)
x_2=np.arange(-3,3,0.1)
X_1,X_2=np.meshgrid(x_1,x_2)
```

```python
X_3 = f(X_1,X_2)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_1,X_2,X_3,cmap=cm.viridis)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("f(x1, x2)")
plt.show()
```

```python
plt.close()
fig,ax=plt.subplots()
ax.set_aspect(aspect=1)
CS=ax.contour(X_1,X_2,X_3,levels=[2,4,6,8,10,15,20,25,30,35,40,50,60,80,100,120,150,200,250,300,400,500])
ax.clabel(CS,inline=1,fontsize=10)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
#ax.plot([-4,2,2],[-1,-1,-4],"--");
```

## Algorithm:
Starting point $x_0 = (1,0)$

```python
f(1,0)
```

```python
grad_f(1,0)
```

```python
hessian_f(1,0)
```

```python
p=[np.array([1,0])]
k=0
while(np.linalg.norm(grad_f(*p[k]))>0.0000001) and k<100:
      v=np.linalg.solve(hessian_f(*p[k]),-grad_f(*p[k])); #Solve for the current point
      p.append(p[k]+v)
      k+=1
      print(f"new point: {p[k]}")
```
